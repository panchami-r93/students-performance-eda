# Introduction & Problem Statement

In this project, I analyze a students performance dataset to understand which factors influence exam scores. The goal is to help schools or coaching centers focus on the factors that matter most, such as study time or test preparation.

## Load Libraries & Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("students_performance.csv")
df.head()
```

 ## Basic Info & Cleaning
 #### Code:

```python
df.info()
df.describe()
df.isna().sum()
```

 ## Univariate Analysis (Single Columns)
- Plot distributions of scores:

```python
sns.histplot(df["math score"], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

```python
sns.histplot(df["reading score"], kde=True)
plt.title("Distribution of Reading Scores")
plt.show()
```

```python
sns.histplot(df["writing score"], kde=True)
plt.title("Distribution of Writing Scores")
plt.show()
```

 ## Bivariate Analysis (Relationships)
 - Example:

```python
sns.boxplot(x="test preparation course", y="math score", data=df)
plt.title("Math Scores vs Test Preparation Course")
plt.show()
```

```python
sns.boxplot(x="gender", y="math score", data=df)
plt.title("Math Scores by Gender")
plt.show()

sns.boxplot(x="race/ethnicity", y="math score", data=df)
plt.title("Math Scores by Race/Ethnicity Group")
plt.show()
```

## Key Findings

- Students who completed the test preparation course tend to have higher math, reading, and writing scores than those who did not.  
- Average scores in all three subjects are above passing level, but there is a wide range between the lowest and highest performers.  
- There are visible differences in scores between race/ethnicity groups, with some groups scoring higher on average than others.  
- The distributions of math, reading, and writing scores are slightly skewed, suggesting more students cluster around certain score ranges.
